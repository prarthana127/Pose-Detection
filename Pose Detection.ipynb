```python
import cv2 
import mediapipe as mp
import numpy as np

mp_drawing=mp.solutions.drawing_utils
mp_pose=mp.solutions.pose
```

# 1. Make Detection

```python
cap=cv2.VideoCapture(0)
#setting up mediapipe instance
with mp_pose.Pose(min_detection_confidence=0.75,min_tracking_confidence=0.75) as pose:
    while cap.isOpened():
        ret,frame=cap.read()
        
        #convert BGR to RGB
        image=cv2.cvtColor(frame,cv2.COLOR_BGR2RGB)
        image.flags.writeable=False

        #Make Detections
        result=pose.process(image)

        #convert RGB to BGR
        image.flags.writeable=True
        image=cv2.cvtColor(image,cv2.COLOR_RGB2BGR)

        #Rendering Results
        mp_drawing.draw_landmarks(image,result.pose_landmarks,mp_pose.POSE_CONNECTIONS,mp_drawing.DrawingSpec(color=(245,117,66),thickness=2,circle_radius=2),
                                                                                mp_drawing.DrawingSpec(color=(245,66,230),thickness=2,circle_radius=2))
        print(result)

        cv2.imshow('MediPipe feed',image)

        if cv2.waitKey(10) & 0xFF==ord('q'):
            break

    cap.release()
    cv2.destroyAllWindows()
    cv2.waitKey(1)
```

# 3. Calculate Angles

```python
def calculate_angle(a, b, c):
    a = np.array(a)  # First point
    b = np.array(b)  # Middle point
    c = np.array(c)  # Last point

    radians=np.arctan2(c[1]-b[1],c[0]-b[0])-np.arctan2(a[1]-b[1],a[0]-b[0])
    angle=np.abs(radians*180/np.pi)

    if angle>180:
        angle=360-angle

    return angle
    
```

# 4. Curl Counter 

```python
cap=cv2.VideoCapture(0)

counter =0;
stage=None

#setting up mediapipe instance
with mp_pose.Pose(min_detection_confidence=0.75,min_tracking_confidence=0.75) as pose:
    while cap.isOpened():
        ret,frame=cap.read()

        #convert BGR to RGB
        image=cv2.cvtColor(frame,cv2.COLOR_BGR2RGB)
        image.flags.writeable=False

        #Make Detections
        result=pose.process(image)
        
        #convert RGB to BGR
        image.flags.writeable=True
        image=cv2.cvtColor(image,cv2.COLOR_RGB2BGR)


        #Extract Landmarks
        try:
            landmarks=result.pose_landmarks.landmark


            #Get coordinates
            shoulder=[landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value].x,landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value].y]
            elbow=[landmarks[mp_pose.PoseLandmark.LEFT_ELBOW.value].x,landmarks[mp_pose.PoseLandmark.LEFT_ELBOW.value].y]
            wrist=[landmarks[mp_pose.PoseLandmark.LEFT_WRIST.value].x,landmarks[mp_pose.PoseLandmark.LEFT_WRIST.value].y]

            #Calculate angle
            angle=calculate_angle(shoulder,elbow,wrist)

            #Visualize angle
            cv2.putText(image, str(angle),
                        tuple(np.multiply(elbow, [640, 480]).astype(int)),
                        cv2.FONT_HERSHEY_COMPLEX, 0.5, (255, 255, 255), 2, cv2.LINE_AA)
    
            if(angle>160):
                stage="down"
            if angle<30 and stage=="down":
                stage="up"
                counter+=1
                
        
        except:
            pass

            cv2.putText(image, f"Count: {counter}", (10, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2, cv2.LINE_AA)
        
        #Rendering Results
        mp_drawing.draw_landmarks(image,result.pose_landmarks,mp_pose.POSE_CONNECTIONS,
                                mp_drawing.DrawingSpec(color=(245,117,66),thickness=2,circle_radius=2),
                                mp_drawing.DrawingSpec(color=(245,66,230),thickness=2,circle_radius=2))


        cv2.imshow('MediPipe feed',image)

        if cv2.waitKey(10) & 0xFF==ord('q'):
            break

    cap.release()
    cv2.destroyAllWindows()
    cv2.waitKey(1)
```

# Squat Detection

```python
cap = cv2.VideoCapture(0)

# Setting up MediaPipe instance
with mp_pose.Pose(min_detection_confidence=0.75, min_tracking_confidence=0.75) as pose:
    while cap.isOpened():
        ret, frame = cap.read()

        # Convert BGR to RGB
        image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        image.flags.writeable = False

        # Make Detections
        result = pose.process(image)
        
        # Convert RGB to BGR
        image.flags.writeable = True
        image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)

        # Extract landmarks for squat detection
        try:
            landmarks = result.pose_landmarks.landmark

            # Get coordinates for key points in squat detection
            hip = [landmarks[mp_pose.PoseLandmark.LEFT_HIP.value].x, landmarks[mp_pose.PoseLandmark.LEFT_HIP.value].y]
            knee = [landmarks[mp_pose.PoseLandmark.LEFT_KNEE.value].x, landmarks[mp_pose.PoseLandmark.LEFT_KNEE.value].y]
            ankle = [landmarks[mp_pose.PoseLandmark.LEFT_ANKLE.value].x, landmarks[mp_pose.PoseLandmark.LEFT_ANKLE.value].y]

            # Calculate angle at the knee
            knee_angle = calculate_angle(hip, knee, ankle)

            # Visualize angle
            cv2.putText(image, f'Knee Angle: {int(knee_angle)}',
                        tuple(np.multiply(knee, [640, 480]).astype(int)),
                        cv2.FONT_HERSHEY_COMPLEX, 0.5, (255, 255, 255), 2, cv2.LINE_AA)

            # Provide feedback based on knee angle
            if knee_angle < 70:
                feedback = "Lower hips more"
            elif knee_angle > 100:
                feedback = "Straighten legs"
            else:
                feedback = "Perfect squat posture!"

            # Display feedback
            cv2.putText(image, feedback,
                        (50, 50),
                        cv2.FONT_HERSHEY_COMPLEX, 0.7, (0, 255, 0), 2, cv2.LINE_AA)

        except:
            pass

        # Rendering Results
        mp_drawing.draw_landmarks(image, result.pose_landmarks, mp_pose.POSE_CONNECTIONS,
                                  mp_drawing.DrawingSpec(color=(245, 117, 66), thickness=2, circle_radius=2),
                                  mp_drawing.DrawingSpec(color=(245, 66, 230), thickness=2, circle_radius=2))

        cv2.imshow('MediPipe Squat Detection', image)

        if cv2.waitKey(10) & 0xFF == ord('q'):
            break

    cap.release()
    cv2.destroyAllWindows()
    cv2.waitKey(1)
```
